```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
df = sns.load_dataset('tips')
df.head()
```

```
df.info()
```

```
pd.set_option('float', '{:.2f}'.format)
df.describe().T
```

```
df.nlargest(5, 'total_bill')
```

```
df.describe(include='category').T
```

```
sns.relplot(data=df, x='total_bill', y='tip')
```

```
sns.relplot(data=df, x='total_bill', y='tip', height=7)
```

```
sns.relplot(data=df, x='total_bill', y='tip', hue='day', size='size', height=7, palette="ch:2.3,-.2,dark=.2")
```

```
sns.relplot(data=df, x='total_bill', y='tip', hue='day', size='size', height=5, col='time', row='smoker', palette="ch:2.3,-.2,dark=.2")
```

```
sns.catplot(data=df, x='day', y='total_bill')
```

```
sns.catplot(data=df, x='day', y='total_bill', kind='swarm')
```

```
sns.catplot(data=df, x='day', y='total_bill', kind='box')
```

```
sns.catplot(data=df, x='day', y='total_bill', kind='violin')
```

```
sns.catplot(data=df, x='day', y='total_bill', kind='bar')
```

```
sns.catplot(data=df, x='day', kind='count')
```
