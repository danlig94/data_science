```
import matplotlib
matplotlib.__version__
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
fig = plt.figure()
fig.show()
```

```
_ = plt.plot([0, 3], [0, 4])
```

```
plt.plot([1, 3], [1, 4], label='line', color='red')
plt.ylabel('axis y')
plt.xlabel('axis x')
plt.title('plot')
_ = plt.legend()
```

# plt.xlabel = 'axis' crash function and need re-import library


```
plt.plot([0, 1, 2], [0, 4, 1], label='line', color='red')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('plot')
_ = plt.legend()
```

```
x = np.arange(-5, 5, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

```
plt.subplot(121)
plt.plot([0, 1, 2], [0, 3, 5])
plt.subplot(122)
_ = plt.plot([0, 1, 2], [0, 3, 5], 'g--')
```

```
fig = plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.plot([0, 1, 2], [0, 3, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [0, 3, 5], 'g--')
plt.plot([0, 1, 2], [1, 4, 3])
plt.subplot(133)
plt.plot([0, 1, 2], [0, 3, 5], 'r.')
```

```
plt.bar(x=[1, 2, 3, 4, 5], height=[1.5, 2.6, 3.5, 1.6, 0.5])
```

```
plt.bar(x=['a', 'b', 'c'], height=[1, 4, 2])
```

```
plt.barh(y=['a', 'b', 'c'], width=[1, 4, 2])
```

```
plt.scatter(x=[1, 2, 3], y=[50, 20, 10])
```

```
_ = plt.pie(x=[35, 5, 10, 25, 25], explode=[0.2, 0.1, 0, 0, 0])
```

```
fig = plt.figure(figsize=(16, 12))

plt.subplot(321)
plt.plot([0, 1, 2], [1, 4, 3])
plt.subplot(322)
plt.plot([3, 4, 1], [1, 4, 3])
plt.subplot(323)
plt.bar(x=[0, 1, 2], height=[1, 4, 3])
plt.subplot(324)
plt.bar(x=[3, 4, 1], height=[1, 4, 3])
plt.subplot(325)
plt.scatter(x=[0, 1, 2], y=[1, 4, 3])
plt.subplot(326)
_ = plt.scatter(x=[3, 4, 1], y=[1, 4, 3])
```
