```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```
fig = plt.figure(figsize=(16, 12))
plt.subplot(221)
plt.imshow(mpimg.imread("business.jpg"))
plt.subplot(222)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
```

```
fig = plt.figure(figsize=(16, 12))
for idx, img in enumerate(['business.jpg', 'children.jpg', 'office.jpg', 'workplace.jpg']):
  plt.subplot(220 + idx + 1)
  plt.imshow(mpimg.imread(img))
```
