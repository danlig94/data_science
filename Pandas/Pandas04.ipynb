```
import pandas as pd
import numpy as np
```

```
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('FB')
df.head()
```

```
df.to_csv('FB.csv')
```

```
df_2020 = df[(df.index >= '2020-01-01') & (df.index < '2021-01-01')]
df_2020
```

```
df_2020.index.size
```

```
df_2020.to_csv('FB_2020')
```

```
new_df = pd.read_csv('FB_2020', index_col=0)
new_df
```
